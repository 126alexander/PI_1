```python
def critical_points(z):

    z_x = diff(z,x)
    z_y = diff(z,y)

    cr_point = solve([z_x, z_y], [x, y], dict=True)

    A = diff(z,x,2)
    B = diff(z,x,y)
    C = diff(z,y,2)

    D = A*C - B**2
    return cr_point, A, D
```

```python
def suff_indic(A, D, cr_point):
    A0 = A.subs(cr_point)
    D0 = D.subs(cr_point)
    return D0, A0
```

```python
x, y = symbols('х у')
z = x**2 - 4*x*y - 2*y**2 + 8*x
cr_point, A, D = critical_points(z)
cr_point
```

# Индивидуальное задание

Паритет опционов пут и колл

$$PV(Call) - PV(Put) = S_0 - PV(K)$$

С денежными потоками в конце срока погашения $T$:

$CF(T, Call) = max(S_T - K, 0) = (S_T - K)^+$

$CF(T, Put)  = max(K - S_T, 0) = (K - S_T)^+$

$CF(T, Fwd) = S_T - K$


Например, если цена опциона колл завышена по сравнению с его реплицирующимся портфелем, т.е. Fwd + Put, продайте колл и купите репликацию. Результирующий сдвиг в спросе и предложении должен привести к корректировке цен.

```python
"""Соотношение спотовой цены и денежного потока (соотношение "пут-колл")"""

import numpy as np
import matplotlib.pyplot as plt

S_T = np.linspace(0,100)
K = 50

# портфолио
CF_T = np.maximum(S_T - K, 0)
CF2_T = S_T - K

# остальное - заговор
plt.figure('PCP', figsize=(9,4))
plt.plot(S_T, CF_T)
plt.plot(S_T, CF2_T)
plt.axhline(y=0, linewidth=1, color="black")
plt.xlabel("Спотовая цена")
plt.ylabel("Денежный поток")
plt.annotate("K", (K,-5))
plt.show()
```
